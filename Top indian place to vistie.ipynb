```python
import pandas as pd
df1=pd.read_csv("Top Indian Places to Visit.csv")
df1
```

```python
col=['Unnamed: 0','Zone','State','City','Name','Establishment Year','time needed to visit in hrs','Entrance Fee in INR','Airport with 50km Radius','Weekly Off','Significance',
    'DSLR Allowed','Best Time to visit']
col
```

```python
df=df1.drop(columns=col)
df
```

```python
x=pd.get_dummies(df,dtype='int')
x
```

```python
y=df1[['Entrance Fee in INR']]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
m1=LinearRegression()
m1.fit(X_train,Y_train)
```

```python
m1.score(X_train,Y_train)
```

```python
from sklearn.linear_model import LogisticRegression
m2=LogisticRegression()
```

```python
m2.fit(X_train,Y_train)
```

```python
m2.score(X_train,Y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
m3= DecisionTreeClassifier()
```

```python
m3.fit(X_train,Y_train)
```

```python
m3.score(X_train,Y_train)
```
